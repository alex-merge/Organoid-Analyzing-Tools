## Computing a full analysis on the dataset

### Looking at the documentation

```python
help(vectors.full_analysis)
```

The full_analysis method allows you to directly compute the vector analysis on the dataframe. It can be usefull if you already have a dataframe that can work with OAT. It basically allow you to bypass the import part. By comparison, this it not possible with the basic use of OAT.
But, you don't have much more control on parameters which can be usefull. For that, you can directly call the methods full_analysis is calling.

Those methods come with a check of what data you are inputing.

### Computing the drift

```python
data = vectors.get_drift(df)
data.head()
```

Here it worked because, in order to compute the drift, you only need the coordinates and the time points.
But if you are trying to compute the angular velocity for example, it needs the rotation axis for each time points, which are not present in data for the moment.

```python
df, data = vectors.get_angular_velocity(df, data)
```
